```python
# Calculate average ratings for each restaurant
restaurant_ratings = df.groupby('Restaurant Name')['Aggregate rating'].mean()

# Count occurrences of each restaurant
restaurant_counts = df['Restaurant Name'].value_counts()

# Combine ratings and counts into a single DataFrame
result_df = pd.DataFrame({'Average Rating': restaurant_ratings, 'Count': restaurant_counts})

# Sort by value counts in descending order
result_df = result_df.sort_values('Count', ascending=False)

# Display the result
print(result_df)
```

```python
(df.groupby('Restaurant Name')['Aggregate rating']
 .agg(['mean', 'count'])
 .rename(columns={'mean': 'Average Rating', 'count': 'Count'})
 .sort_values('Count', ascending=False))
```

```python
(df.groupby('Restaurant Name')['Aggregate rating']
 .agg(['mean', 'count'])
 .rename(columns={'mean': 'Average Rating', 'count': 'Count'})
 .sort_values('Average Rating', ascending=False))
```

```python
(df[['Restaurant Name','Votes']].sort_values(by='Votes',ascending=False).head(14))
```

```python
# Sort the DataFrame by 'Votes' and select the top 10
top_10_restaurants = df[['Restaurant Name', 'Votes']].sort_values(by='Votes', ascending=False).head(10)

# Create the bar plot
plt.figure(figsize=(10, 6))  # Adjust figure size as needed
sns.barplot(x='Restaurant Name', y='Votes', data=top_10_restaurants)
plt.xticks(rotation=90)  # Rotate x-axis labels for readability
plt.title('Top 10 Restaurants by Votes')
plt.xlabel('Restaurant Name')
plt.ylabel('Number of Votes')
plt.tight_layout()  # Adjust layout to prevent labels from overlapping
plt.show()
```
